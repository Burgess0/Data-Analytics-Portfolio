# 本程序是基于jieba分词的全文本词频统计

## 全文本词频统计的步骤
+ 打开文本
+ 分词
+ 去除停用词（集合的方式）
+ 利用字典，进行词-词频的存储
+ 排序
+ 展示（输出print）
+  
+ 这个是非常简单的python程序

```python
import jieba
```

```python
article = open('sanguo_10.txt','r',encoding = 'ANSI').read() # 打开并读取三国前10回 #出现乱码提示，就把ANSI改成utf-8
```

```python
dele = {'。','！','？','的','“','”','（','）',' ','》','《','，'} # 手动设计一些停用词和符号
```

```python
jieba.add_word('老贼') # 加入字典中没有的新词
```

```python
words = list(jieba.cut(article)) #结巴分词出来的词汇
```

```python
words
```

### 字典

```python
articleDict = {} # 这是一个字典，准备词-词频的保存
```

### 集合（数据结构的一种）

```python
articleSet = set(words)-dele
```

```python
articleSet
```

```python
# 因为采用的是循环的方法，把所有的词都循环一遍，长度大于1，所以速度很慢！
```

```python
for w in articleSet:
    if len(w)>1:
        articleDict[w] = words.count(w)  # 字典的key-value配对
```

```python
articleDict
```

```python
articlelist = sorted(articleDict.items(),key = lambda x:x[1], reverse = True) # 对词典中的词排序
```

```python
# 输出词频的前N个
for i in range(100):
    print(articlelist[i])
```
