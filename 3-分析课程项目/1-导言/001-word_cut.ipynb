载入停用词表

```python
stopwords = [line.strip() for line in open('stop_words.txt', 'r', encoding='utf-8').readlines()]  
```

```python
seg_list = jieba.cut("使用了停用词表之后啊，效果就好看很多了，什么啊、了、是之类的词就不见了")
```

```python
final = ''
```

```python
for seg in seg_list:
    if seg not in stopwords:
        final += seg+'*'
```

```python
print (final)   # 做实体抽取的时候，停用词表很管用
```
